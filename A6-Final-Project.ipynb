RCW 42.17A.001(10)  
  
Following the passage of the Federal Election Campaign Act of 1971 by the United States Congress, Washington state voters approved the Campaign Disclosure and Contribution statute in 1972, codified in Revised Code of Washington (RCW) chapter 42.17A. These laws require regular reporting of political contributions and set contribution limitations. For more information, see the Washington Attorney General (WA AG) [Enforcement of Campaign Finance Laws](http://www.atg.wa.gov/enforcement-campaign-finance-laws) website.  
   
RCW 42.17A also created Washington's five member Public Disclosure Commission (PDC). The purpose of the PDC is "to provide timely and meaningful public access to accurate information about the financing of political campaigns, lobbyist expenditures, and the financial affairs of public officials and candidates, and to ensure compliance with and equitable enforcement of Washington's disclosure and campaign finance laws." [PDC website](https://www.pdc.wa.gov/learn) Specifically, the PDC must operate a web site that allows public access to the campaign finance data that must be disclosed. RCW 42.17A.050.   
  
Contribution Limits are listed on the [PDC website](https://www.pdc.wa.gov/learn/contribution-limits) The limits vary based on the identity of both the recipient and the contributor, and whether a candidate raises and spends less than \$5,000.
```python
# Helper functions

col_list = ["id", "report_number", "origin", "first_name", "middle_initial",
            "last_name", "description", "code", "memo",
            "description", "contributor_employer_city"]

def drop_cols(df, cols=col_list):
    # Drops the same columns for each dataframe read from CSV file
    try:
        for col in df.columns:
            if col in cols:
                print("Dropped Column:", col)
                df.drop(col, axis=1, inplace=True)
    except:
        print("No columns dropped")

def highlight_max(data, color='yellow'):
    '''
    http://pandas.pydata.org/pandas-docs/stable/style.html
    highlight the maximum in a Series or DataFrame
    Series (from .apply(axis=0 or 1)) or DataFrames (from .apply(axis=None))
    Example highlighting max value in 3 columns of df:
      df.style.apply(highlight_max, subset=['B', 'C', 'D'])
    Example:  df.style.apply(highlight_max)
      uses .apply which operates columnwise (or rowwise w/ the axis keyword)
    Example for highlighting max value in df w/ axis=None:
      df.style.apply(highlight_max, color='darkorange', axis=None)
    '''
    attr = 'background-color: {}'.format(color)
    if data.ndim == 1:  # Series from .apply(axis=0) or axis=1
        is_max = data == data.max()
        return [attr if v else '' for v in is_max]
    else:  # from .apply(axis=None)
        is_max = data == data.max().max()
        return pd.DataFrame(np.where(is_max, attr, ''),
                            index=data.index, columns=data.columns)
    
```

### Data Preparation

Using python, I have read the text file of the dataset into a pandas dataframe and conducted some preliminary descriptive statistics. I don't plan to use the id, report_number, origin, first_name, middle_initil, last_name, memo, or url columns. Given the large dataset, I plan to create several pandas dataframes with subsets of data. At this point, breaking the data up by election year and type of campaign seems reasonable.  

### Analytical Methods  

Much of the analysis will involve grouping and filtering subsets of the data. Then, functions like counting, averaging, or finding maximum values across a particular feature set will be necessary to address the research questions. Plotting the data will also be important.  

Finally, machine learning algorithms like k means, k nearest neighbor (k-nn) or histogram-based outlier score may be used for trend and anomaly detection. However, since the complexity of k-nn is O$(n^2)$, the input data will need to be carefully curated.  

To that end, principal component analysis (PCA) may be used to detect subspaces in the dataset, like which features tend to have the most variation. The principal components are the eigenvectors of the covariance matrix, so depending on the number of outliers, a robust PCA method that computes the covariance matrix twice based on the Mahalanobis distance may be used.
