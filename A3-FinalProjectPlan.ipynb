<h2 id="i1">I. Introduction</h2>   
__Why are you planning to do this analysis?__  
  
This project will analyze the publicly disclosed contributions to local, meaning not federal, political campaigns and committees. 

Campaign finance is an import issue in any democracy. Citizens "have the right to expect from their elected representatives at all levels of government the utmost of integrity, honesty, and fairness in their dealings." RCW 42.17A.001(2). "[T]he public's right to know of the financing of political campaigns and lobbying and the financial affairs of elected officials and candidates far outweighs any right that these matters remain secret and private." RCW 42.17A.001(10)  

Following the passage of the Federal Election Campaign Act of 1971 by the United States Congress, Washington state voters approved the Campaign Disclosure and Contribution statute in 1972, codified in Revised Code of Washington (RCW) chapter 42.17A. These laws govern the state’s campaign finance requirements by, among other things, requiring campaign finance reporting and setting contribution limits. [Enforcement of Campaign Finance Laws](http://www.atg.wa.gov/enforcement-campaign-finance-laws)   

Further underscoring the importance of this issue, the Campaign Disclosure and Contribution statute created two enforcement mechanisms. First, RCW 42.17A gives the Washington State Attorney General (WA AG) authority for investigating alleged violations of the campaign finance laws and bringing civil suits against offenders. The WA AG website includes a list of recent lawsuits: 9 in 2017 (so far), 5 in 2016, and then no lawsuits until 2011. Either that list is incomplete, or the WA AG hasn't filed enough lawsuits because there's no way those cases represent the majority of campaign finance violations. 

Second, RCW 42.17A created the five member Public Disclosure Commission (PDC). The PDC hears alleged campaign violations in a quasi-judicial proceeding, but can only impose a maximum civil penalty of $10,000. The purpose of the PDC is "to provide timely and meaningful public access to accurate information about the financing of political campaigns, lobbyist expenditures, and the financial affairs of public officials and candidates, and to ensure compliance with and equitable enforcement of Washington's disclosure and campaign finance laws." [PDC website](https://www.pdc.wa.gov/learn) Specifically, the PDC must operate a web site that allows public access to the campaign finance data that must be disclosed. RCW 42.17A.050.

Although, "contribution and expenditure limitations operate in an area of the most fundamental First Amendment activities." _Buckley v. Valeo_, 424 U.S. 1, 14 (1976). The Court ruled that, while contributions could be limited in order to avoid corruption, or the appearance of corruption, spending by individuals or groups or by candidates themselves (from their own personal resources) could not corrupt elections and should not be limited under the First Amendment. Subsequent decisions have further weakened campaign finance regulations. For example, _Citizens United v. Federal Election Comm'n_, 558 U.S. 310 (2010) held that restrictions on the amount or type of corporate spending in politics constitutes a prior restraint in violation of the First Amendment, but that disclaimer and disclosure requirements are valid since they are less restrictive. Then, a later decision held that the _Citizens United_ decision applies to both federal and state elections.

### Analytical Methods  

Much of the analysis will involve grouping and filtering subsets of the data. Then, functions like counting, averaging, or finding maximum values across a particular feature set will be necessary to address the research questions. Plotting the data will also be important.  

Finally, machine learning algorithms like k means, k nearest neighbor (k-nn) or histogram-based outlier score may be used for trend and anomaly detection. However, since the complexity of k-nn is O$(n^2)$, the input data will need to be carefully curated.  

To that end, principal component analysis (PCA) may be used to detect subspaces in the dataset, like which features tend to have the most variation. The principal components are the eigenvectors of the covariance matrix, so depending on the number of outliers, a robust PCA method that computes the covariance matrix twice based on the Mahalanobis distance may be used.
